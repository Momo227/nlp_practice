## バイアス
* バイアスを含めることで計算の柔軟性が高まる。

####・バイアスがない場合
  * 活性化関数への入力は単に$xw$である。
  * 単純なネットワークで重み$w$変更するのかをどのようなことが起きるかを見る。

```python
import matplotlib.pylab as plt
import numpy as np

x = np.arange(-6, 6, 0.1)
weights = [0.5, 1.0, 2.0]
for i, w in enumerate(weights):
    f = 1/ (1+np.exp(-x*w))
    label = 'w{} = {}'.format(i, w)
    plt.plot(x, f, label=label)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2)
plt.show()
```

↑のグラフより、重みを変えることでシグモイド関数の出力の傾きが変わることが確認できる。

→入力変数と出力変数の間の関係の強さをモデル化したい場合に便利。

 #### ・活性化関数への入力が$xw+b$の場合。

```python
import matplotlib.pylab as plt
import numpy as np

x = np.arange(-6, 6, 0.1)
w = 5
biases = [-8.0, 0.0, 8.0]
for i, b in enumerate(biases):
    f = 1/ (1+np.exp(-x*w+b))
    label = 'b{} = {}'.format(i, b)
    plt.plot(x, f, label=label)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2)
plt.show()
```

↑バイアスの値ごと(-8.0, 0.0, 8.0)に関数が表示されていることが確認できる。

バイアス$b$を変化させることで、ノードがアクティブになるタイミングを変えることができる。

→よって、バイアスを追加することで、入力$x$がある閾値以上だったら出力を1、そうでないなら0というような関係を表現できる。
